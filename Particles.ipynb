```python
import matplotlib.pyplot as plt
import math

# Dimensiones de la caja
box_width = 100
box_height = 80
# Número de círculos
num_circles = 20
# Radio fijo de los círculos
circle_radius = 5

# Calcular el número de filas y columnas
columns = int(math.sqrt(num_circles * (box_width / box_height)))
rows = math.ceil(num_circles / columns)

# Calcular separación entre círculos
horizontal_spacing = (box_width - 2 * circle_radius) / (columns - 1) if columns > 1 else 0
vertical_spacing = (box_height - 2 * circle_radius) / (rows - 1) if rows > 1 else 0

# Generar los círculos distribuidos uniformemente
circles = []
for row in range(rows):
    for col in range(columns):
        if len(circles) >= num_circles:
            break
        x_position = circle_radius + col * horizontal_spacing
        y_position = circle_radius + row * vertical_spacing
        circles.append((x_position, y_position, circle_radius))

# Visualizar los círculos
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, box_width)
ax.set_ylim(0, box_height)
ax.set_aspect('equal')
ax.set_title("20 Círculos azules distribuidos uniformemente")

# Dibujar los círculos
for x, y, r in circles:
    circle = plt.Circle((x, y), r, edgecolor='blue', facecolor='blue', alpha=0.7)
    ax.add_patch(circle)

plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
import random
import math

# Dimensiones de la caja
box_width = 100
box_height = 80
# Número de círculos
num_circles = 20
# Radio fijo de los círculos
circle_radius = 5
# Lista para almacenar los círculos generados
circles = []

def are_circles_overlapping(x1, y1, r1, x2, y2, r2):
    """Comprueba si dos círculos se solapan."""
    distance = math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
    return distance < r1 + r2

def generate_circle():
    """Genera un círculo aleatorio dentro de la caja."""
    while True:
        x = random.uniform(circle_radius, box_width - circle_radius)
        y = random.uniform(circle_radius, box_height - circle_radius)
        
        # Comprobar si se solapa con otros círculos
        overlap = False
        for cx, cy, cr in circles:
            if are_circles_overlapping(x, y, circle_radius, cx, cy, cr):
                overlap = True
                break

        if not overlap:
            return x, y

# Generar los círculos
for _ in range(num_circles):
    x, y = generate_circle()
    circles.append((x, y, circle_radius))

# Visualizar los círculos
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, box_width)
ax.set_ylim(0, box_height)
ax.set_aspect('equal')
ax.set_title("20 Particles randomly placed")

# Dibujar los círculos
for x, y, r in circles:
    circle = plt.Circle((x, y), r, edgecolor='blue', facecolor='blue', alpha=0.7)
    ax.add_patch(circle)

plt.grid(True)
plt.show()
```
